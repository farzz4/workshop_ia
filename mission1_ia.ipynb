Etape 0 : Les imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

Etape 1 : Charger les données

```python
df = pd.read_csv("visiteurs.csv")
df
```

```python
donnees = df.values
donnees
```

Etape 2 : Déclarer les features (X)

Les features :
Colonne 1 → Age              
Colonne 2 → Taille_Groupe    
Colonne 3 → Nb_Enfants       
Colonne 4 → Score_Physique

```python
X_train = donnees[:, [1, 2, 3, 4]]
X_train
```

Etape 3 : Entraîner le modèle

```python
modele = KMeans(n_clusters=3)
modele.fit(X_train)
```

```python
y_train_pred = modele.labels_
print("Groupes attribués aux visiteurs :")
print(y_train_pred)
```

Etape 4 : Afficher les centres

```python
centres = modele.cluster_centers_
centres
```

```python
print("Centres des groupes :")
for i, c in enumerate(centres):
    print(f"  Groupe {i} → âge moyen={c[0]:.1f}, taille groupe={c[1]:.1f}, nb enfants={c[2]:.1f}, score physique={c[3]:.1f}")
```

Etape 5 : Prédire le groupe pour de nouveaux visiteurs

```python
X_test = np.array([
    [23, 1, 0, 9],   # Visiteur A : jeune, seul, sans enfants, très actif
    [38, 4, 2, 3],   # Visiteur B : parent, grande famille, peu actif
    [44, 2, 1, 4],   # Visiteur C : adulte, petit groupe, modéré
])
X_test
```

```python
y_test_pred = modele.predict(X_test)
print("Groupes prédits pour les nouveaux visiteurs :")
for i, (visiteur, groupe) in enumerate(zip(X_test, y_test_pred)):
    print(f"  Visiteur {chr(65+i)} → Groupe {groupe}")
```

Etape 6 : Visualisation des clusters

```python
cluster0 = X_train[y_train_pred==0]
cluster1 = X_train[y_train_pred==1]
cluster2 = X_train[y_train_pred==2]
```

```python
plt.plot(cluster0[:,0], cluster0[:,1], 'ro', label='Familles')
plt.plot(cluster1[:,0], cluster1[:,1], 'go', label='Jeunes Aventuriers')
plt.plot(cluster2[:,0], cluster2[:,1], 'bo', label='Adultes Tranquilles')
plt.xlabel('Age')
plt.ylabel('Taille Groupe')
plt.legend()
plt.show()
```
