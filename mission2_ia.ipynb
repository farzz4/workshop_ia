Etape 0 : Importer les bibliothèques

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

Etape 1 : Charger les données

```python
df = pd.read_csv("jours.csv")
df
```

```python
donnees = df.values
donnees
```

Etape 2 : Séparer les features (X) et la cible (y)

Colonne 1 → Est_Weekend        -> feature  
Colonne 2 → Est_Ferie          -> feature
Colonne 3 → Temperature        -> feature
Colonne 4 → Precipitation      -> feature
Colonne 5 → Vacance_Scolaire   -> feature
Colonne 6 → Nb_Visiteurs       -> cible

```python
X = donnees[:, [1, 2, 3, 4, 5]]
X
```

```python
y = donnees[:, 6]
y
```

Etape 3 : Split 70% Train / 30% Test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Les features utilisées pour apprendre
print("X_train (features pour apprendre) :")
print(X_train)

# Les features utilisées pour tester
print("\nX_test (features pour tester) :")
print(X_test)

# Les vraies réponses pour apprendre
print("\ny_train (nb visiteurs réels pour apprendre) :")
print(y_train)

# Les vraies réponses qu'on va comparer aux prédictions
print("\ny_test (nb visiteurs réels pour tester) :")
print(y_test)
```

Etape 4 : Entaîner le modèle

```python
modele = LinearRegression()
modele.fit(X_train, y_train)
```

```python
print("Modèle entraîné !")
print(f"\nCoefficients   : {modele.coef_}")
print(f"Ordonnée orig. : {modele.intercept_:.1f}")
```

Etape 5 : Prédire pour X_test

```python
y_pred = modele.predict(X_test)

print("y_test :", y_test)
print("y_pred :", y_pred)
```

Etape 6 : Évaluer le modèle avec RMSE et R²

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"RMSE : {rmse:.1f} visiteurs  → en moyenne, le modèle se trompe de {rmse:.0f} visiteurs")
print(f"R²   : {r2:.2f}             → le modèle explique {r2*100:.0f}% de la variabilité")
```
